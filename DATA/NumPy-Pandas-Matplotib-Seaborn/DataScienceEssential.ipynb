# Combining and Merging DataFrames

why need to merge ?
- lets say we have population dataset of different years, and you want to have a single dataset combining all of them , it is one of them among many others reasons.

#Methods to do it
- Concatenation
- Merging
- Joining

- Rowwise Combining (axis=0)

```python
# lest create two sample data set;

product1  = {
    "name" : [" apple","banana","papaya"],
    "price":["45/kg",'10/pcs',"50/kg"]
}
df1 = pd.DataFrame(product1)
product2  = {
    "name" : [" Mango","Litchi","Guava"],
    "price":["45/kg",'10/pcs',"50/kg"]
}
df2 = pd.DataFrame(product2)

combined_product_dataset = pd.concat([df1,df2],axis=0)
combined_product_dataset
```

- we have redundant index lets remove that using ignore_index=True Parameter

```python
combined_product_dataset = pd.concat([df1,df2],axis=0,ignore_index=True)
combined_product_dataset
```

- column wise combining axis=1

```python
# lest create two sample data set;

product1  = {
    "name" : [" apple","banana","papaya"],
    "price":["45/kg",'10/pcs',"50/kg"]
}
df1 = pd.DataFrame(product1)
product2  = {
    "name" : [" Mango","Litchi","Guava"],
    "price":["45/kg",'10/pcs',"50/kg"]
}
df2 = pd.DataFrame(product2)

combined_product_dataset = pd.concat([df1,df2],axis=1)
combined_product_dataset
```

# Data Aggregation and Grouping


why grouping data ?
- it allows to perform opration on subsets of data based on shared category .
- we can use groupby in Pandas
- SYNTAX:
- grouped = df.groupby("column_name")

  #operation we can perform
    - iterates over groups


```python
import pandas as pd

df = pd.DataFrame({
    "id":[1,2,3,4,5,6,7],
    "department": ["IT", "IT", "HR", "HR", "Finance", "Finance", "IT"],
    "name": ["Aman", "Sita", "Rahul", "Neha", "Kiran", "Rohit", "Priya"],
    "salary": [50000, 60000, 45000, 47000, 70000, 72000, 65000],
    "experience": [2, 3, 1, 2, 5, 6, 4]
})

grouped = df.groupby("id")
for name, group in grouped:
    print(group)
```

- Aggregations functions

   - using group by


```python
df.groupby("department")   ["salary"].mean()
                           # numeric column
```

- power of grouby , using it you can combine all mean max min etc altogether

```python
df.groupby("department").agg({"salary":["mean","max","min"]})
```

```python
# another way
pivot = df.pivot_table(
    values="salary",
    index="department",
    aggfunc="mean"
)
pivot
```

- custom aggregation

```python
def range_func(x):
    return x.max()-x.min()
df.groupby("department")["salary"].agg(range_func)
```

# calculating summary statistics for grouped data
 - common statistics
    - mean
    - max
    - min

```python
df.groupby("department")["salary"].mean()
df.groupby("department")["salary"].max()
df.groupby("department")["salary"].min()
```

```python
df.groupby("department").agg(
    {
        "salary":["mean","max","min"]
    }
)
```
